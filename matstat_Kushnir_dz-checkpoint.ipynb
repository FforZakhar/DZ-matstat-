```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from scipy.stats import t, chi2, f, norm, poisson, binom, uniform, multivariate_normal

# задание 1


np.random.seed(123)

def ecdf(c, x):
    return np.mean(x <= c)



# Хи-Квадрат с 5-ю степенями свободы

def Xi_Dist(n, df):
    x = chi2.rvs(size = n, df = df)

    x = np.sort(x)

    F_x = chi2.cdf(x, df = df)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.plot(x, F_x, '--', markersize = 1,
    label = "CDF", color = "limegreen")
    seaborn.ecdfplot(x, label = "ECDF", color = "palevioletred")
    plt.legend()
    plt.show()

def Xi_Dens(n, df, b):
    x = chi2.rvs(size=n, df=df)
    x = np.sort(x)
    f_x = chi2.pdf(x, df = df)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(x, f_x, '--', markersize = 20,
    label = "PDF", color = "limegreen")
    seaborn.histplot(x, stat = 'density', color = "palevioletred", label = "histogram", bins = b)
    plt.legend()
    plt.show()

n = 1000

Xi_Dist(n, 5)
Xi_Dens(n, 5, 100)
# Стьюдента с 10-ю степенями свободы


def St_Dist(n, df):
    x = t.rvs(size = n, df = df)
    x = np.sort(x)
    F_x = t.cdf(x, df = df)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.plot(x, F_x, '--', markersize = 1,
      label = "CDF", color = "red")
    seaborn.ecdfplot(x, label = "ECDF", color = "green")
    plt.legend()
    plt.show()

def Xi_Dens(n, df, b):
    x = t.rvs(size=n, df=df)
    x = np.sort(x)
    f_x = t.pdf(x, df = 10)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(x, f_x, '--', markersize = 20,
      label = "PDF", color = "blue")
    seaborn.histplot(x, stat = 'density', color = "palevioletred", label = "histogram", bins = b)
    plt.legend()
    plt.show()

St_Dist(n, 10)
Xi_Dens(n, 10, 100)

# Нормальное с математическим ожиданием 10 и дисперсией 25

def Norm_Dist(n, e, d):
    x = norm.rvs(size = n, loc = e , scale = d)
    x = np.sort(x)
    F_x = norm.cdf(x, loc = e , scale = d)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.plot(x, F_x, '--', markersize = 1,
      label = "CDF", color = "red")
    seaborn.ecdfplot(x, label = "ECDF", color = "green")
    plt.legend()
    plt.show()

def Norm_Dens(n, e, d, b):
    x = norm.rvs(size=n, loc=e, scale=d)
    x = np.sort(x)
    f_x = norm.pdf(x, loc = e , scale = d)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.plot(x, f_x, '--', markersize = 20,
      label = "PDF", color = "blue")
    seaborn.histplot(x, stat = 'density', color = "palevioletred", label = "histogram", bins = b)
    plt.legend()
    plt.show()



Norm_Dist(n, 10, 25)
Norm_Dens(n, 10, 25, 100)


#задание 2

sizes = [10, 100, 1000, 10000]

for i in sizes:
    Xi_Dist(i, 5)
    Xi_Dens(i, 10, i//5)
    St_Dist(n, 10)
    Xi_Dens(n, 10, i//5)
    Norm_Dist(i, 10, 25)
    Norm_Dens(i, 10, 25, i//5)



#задание 3

def distrib_func(x, X):
    # x - точка в которой считаем значение
    # X - массив, содержащий выборку
    I = 0
    for i in X:
        if i <= x:
            I+=1
    n = len(X)
    res = I/n
    return res


#задание 4

def hist_func(x, X, b):
    # x - точка в которой считаем значение
    # X - массив, содержащий выборку
    # b - число столбцов в гистограмме
    min_bruh = min(X)
    max_bruh = max(X)
    width = (max_bruh - min_bruh) / b
    n = len(X)
    I = 0
    h1 = max_bruh
    h2 = min_bruh
    for j in range(0, b):
        if min_bruh + width * j <= x <= min_bruh + width * (j + 1):
            h1 = width * j
            h2 = width * (j + 1)
    for i in X:
        if h1 < i < h2:
            I += 1

    res = I / (width * n)
    return res


# задание 5

def distrib_bruh_graph(x):
    x = np.sort(x)
    plt.xlabel('x')
    plt.ylabel('F(x)')
    seaborn.ecdfplot(x,  label="ECDF", color="palevioletred")
    plt.legend()
    plt.show()


def hist(x, b):
    x = np.sort(x)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    seaborn.histplot(x, stat='density', color="red", label="histogram", bins=b)
    plt.legend()
    plt.show()




# задание 6


data = list()


for i in range(300, 333):
    np.random.seed(i)
    x = chi2.rvs(size=100, df=5)
    a = distrib_func(5, x)
    data.append(a)

print(np.mean(data))


print(np.var(data))

```

```python
# задание 1

xi = chi2.rvs(size = 100000, df = 2)

st = t.rvs(size = 100000, df = 2)

norm = norm.rvs(size = 100000, loc = 10 , scale = 25)

# задание 2
m = 1000

x_exs = np.empty(m)
x_exs3 = np.empty(m)
x_vars = np.empty(m)
x_meds = np.empty(m)
x_ps = np.empty(m)
x_quans = np.empty(m)
x_varsins = np.empty(m)

def ecdf(c, x):
    return np.mean(x <= c)

for i in range(0, m):
    x = chi2.rvs(size=200, df=10)

    a = np.mean(x)
    x_exs[i] = a

    b = np.mean(x**3)
    x_exs3[i] = b

    c = np.var(x)
    x_vars[i] = c

    d = np.median(x)
    x_meds[i] = d

    e = ecdf(1.7, x)
    x_ps[i] = e

    f = np.quantile(x, 0.3)
    x_quans[i] = f

    j = np.var(np.sin(x))
    x_varsins[i] = j

print(np.mean(x_exs))
print(np.mean(x_exs3))
print(np.mean(x_vars))
print(np.mean(x_meds))
print(np.mean(x_ps))
print(np.mean(x_quans))
print(np.mean(x_varsins))


# задание 3

m = 1000

e1 = np.empty(m)
e2 = np.empty(m)
e3 = np.empty(m)

for i in range(0, m):
    x = uniform.rvs(size=10000, loc=1, scale=5)
    y = chi2.rvs(size=10000, df=5)
    a = np.mean(x+y)
    e1[i] = a
    b = ecdf(2, x - y)
    e2[i] = b
    c = np.var(x/y)
    e3[i] = c


print(np.mean(e1))
print(np.mean(e2))
print(np.mean(e3))


# задание 4

l = 5

n = 1000

def q_X(p, l):
    e = -(np.log(1-p))/l
    return e

def l_1(x):
    return np.mean(x)

def l_2(x):
    return 1 / np.mean(x)

def l_3(x):
    return 1 / (10 + np.mean(x))

def l_4(x):
    return  1 / (np.log(n) + np.mean(x))

def l_5(x):
    return 1 / (np.log(n) / n + np.mean(x))

def l_6(x):
    return 1 / np.median(x)

def l_7(x):
    return np.log(2) / np.median(x)

l_real = l


l_1_vec = np.empty(n)
l_2_vec = np.empty(n)
l_3_vec = np.empty(n)
l_4_vec = np.empty(n)
l_5_vec = np.empty(n)
l_6_vec = np.empty(n)
l_7_vec = np.empty(n)

for i in range(0, n):
    p_new = np.random.uniform(size = i + 1)
    x_new = q_X(p_new, l)
    l_1_vec[i] = l_1(x_new)
    l_2_vec[i] = l_2(x_new)
    l_3_vec[i] = l_3(x_new)
    l_4_vec[i] = l_4(x_new)
    l_5_vec[i] = l_5(x_new)
    l_6_vec[i] = l_6(x_new)
    l_7_vec[i] = l_7(x_new)


plt.xlabel('Sample Size')
plt.ylabel('Estimate of lambda')
plt.plot(l_1_vec, 'ro', markersize = 2, color = "red", label = "l_1")
plt.plot(l_2_vec, 'ro', markersize = 2, color = "blue", label = "l_2")
plt.plot(l_3_vec, 'ro', markersize = 2, color = "green", label = "l_3")
plt.plot(l_4_vec, 'ro', markersize = 2, color = "yellow", label = "l_4")
plt.plot(l_5_vec, 'ro', markersize = 2, color = "gray", label = "l_5")
plt.plot(l_6_vec, 'ro', markersize = 2, color = "orange", label = "l_6")
plt.plot(l_7_vec, 'ro', markersize = 2, color = "pink", label = "l_7")
plt.axhline(y = l , linestyle='-', color='cyan', label = "lambda")
plt.ylim(ymin = 0, ymax = 12)
plt.legend()
plt.show()








l = 2

n = 300

def Poi(k, l):
    po = poisson.rvs(mu = l, size = k)
    return po

def l_1(x):
    return np.mean(x)

def l_2(x):
    return np.var(x)


def l_4(x):
    return  np.median(x)

def l_5(x):
    return np.median(x) + (1 / 3)

l_real = l


l_1_vec = np.empty(n)
l_2_vec = np.empty(n)
# l_3_vec = np.empty(n)
l_4_vec = np.empty(n)
l_5_vec = np.empty(n)


for i in range(0, n):
    # p_new = np.random.uniform(size = i + 1)
    x_new = Poi(i+1, l)
    l_1_vec[i] = l_1(x_new)
    l_2_vec[i] = l_2(x_new)
    # l_3_vec[i] = l_3(x_new)
    l_4_vec[i] = l_4(x_new)
    l_5_vec[i] = l_5(x_new)


plt.xlabel('Sample Size')
plt.ylabel('Estimate of lambda')
plt.plot(l_1_vec, 'ro', markersize = 2, color = "red", label = "l_1")
plt.plot(l_2_vec, 'ro', markersize = 2, color = "blue", label = "l_2")
# plt.plot(l_3_vec, 'ro', markersize = 2, color = "green", label = "l_3")
plt.plot(l_4_vec, 'ro', markersize = 2, color = "yellow", label = "l_4")
plt.plot(l_5_vec, 'ro', markersize = 2, color = "black", label = "l_5")
plt.axhline(y = l , linestyle='-', color='cyan', label = "lambda")
# plt.ylim(ymi, ymax = )
plt.legend()
plt.show()







rho = 0.5
n = 200

means = [0, 0]
cov = [[1, rho],
       [rho, 1]]


def Z_N(means, cov, k ):
    z= multivariate_normal.rvs(mean = means, cov = cov, size = k)
    x = np.empty(k)
    y = np.empty(k)
    for i in range(k):
        x[i] = z[i][0]
        y[i] = z[i][1]
    return x, y


def rho_1(x, y):
    return np.correlate(x, y)

def rho_2(x, y):
    return np.mean(x * y)

def rho_3(x, y):
    return 0.2 * np.correlate(x, y) + 0.5 * np.mean(x * y)

rho_1_vec = np.empty(n)
rho_2_vec = np.empty(n)
rho_3_vec = np.empty(n)

for i in range(4, n):
    x_new = Z_N(means, cov, i)[0]
    y_new = Z_N(means, cov, i)[1]
    rho_1_vec[i] = rho_1(x_new, y_new)
    rho_2_vec[i] = rho_2(x_new, y_new)
    rho_3_vec[i] = rho_3(x_new, y_new)

plt.xlabel('Sample Size')
plt.ylabel('Estimate of lambda')
plt.plot(rho_1_vec, 'ro', markersize = 2, color = "red", label = "rho_1")
plt.plot(rho_2_vec, 'ro', markersize = 2, color = "blue", label = "rho_2")
plt.plot(rho_3_vec, 'ro', markersize = 2, color = "green", label = "rho_3")
plt.axhline(y = rho , linestyle='-', color='cyan', label = "rho")
# plt.ylim(ymi, ymax = )
plt.legend()
plt.show()








```
